# Improve Image Contrast 

```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
```

Low contrast images typically have a narrow pixel value distribution; i.e., the standard deviation of pixel values is small

```python
img_path = '../Images/Traffic/segment-12304907743194762419_1522_000_1542_000_with_camera_labels_2.png'
```

```python
img = Image.open(img_path)
```

For example, this image has fairly low contrast

```python
img
```

Plot the pixel value of each channel to see the distribution:

```python
def plot_histogram(img):
    """ plot channel-wise pixel value distribution """
    histogram = img.histogram()

    R = histogram[0:256]
    G = histogram[256:512]
    B = histogram[512:768]

    plt.plot(range(256), R, color='r')
    plt.fill_between(range(256), R, color='r', alpha=0.5)
    plt.plot(range(256), G, color='g')
    plt.fill_between(range(256), G, color='g', alpha=0.5)
    plt.plot(range(256), B, color='b')
    plt.fill_between(range(256), B, color='b', alpha=0.5)
    plt.show()
```

```python
plot_histogram(img)
```

Use Pillow Image Enhance to increase the contrast

```python
enhancer = ImageEnhance.Contrast(img)
```

```python
img_enhanced = enhancer.enhance(1.5)
```

```python
img_enhanced
```

```python
plot_histogram(img_enhanced)
```
